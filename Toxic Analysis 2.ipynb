```python
like_stats = df.groupby(['country_map', 'Overall_toxic'])['like_count'].mean().unstack()
like_stats.columns = ['Non-Toxic Average Like', 'Toxic Average Like']
like_stats = like_stats.sort_values(by='Toxic Average Like', ascending=False)
like_stats.columns = ['Non-Toxic', 'Toxic']

country_codes = {
    'Armenia': 'AM', 'Bangladesh': 'BD', 'Catalonia': 'CT', 'China': 'CN',
    'Cuba': 'CU', 'Honduras': 'HN', 'Indonesia': 'ID',
    'Iran': 'IR', 'Mexico': 'MX', 'Russia': 'RU', 'Saudi': 'SA', 
    'Serbia': 'RS', 'Spain': 'ES', 'Tanzania': 'TZ', 'Turkey': 'TR',
    'UAE': 'AE', 'Uganda': 'UG', 'Venezuela': 'VE'
}


like_stats = like_stats.copy()
like_stats['country_code'] = like_stats.index.map(lambda x: country_codes.get(x.strip(), ''))
like_stats = like_stats[like_stats['country_code'] != '']
like_stats.set_index('country_code', inplace=True)
like_stats.sort_index(inplace=True)

fig, ax = plt.subplots(figsize=(14, 6), dpi=600) 
like_stats.plot(kind='bar', width=0.7, color=['#4daf4a', '#e41a1c'], logy=True, ax=ax)
ax.set_xlabel('Country Code', fontsize=16)
ax.set_ylabel('Log-scaled Average Like Count', fontsize=16)
ax.set_xticks(range(len(like_stats)))
ax.set_xticklabels(like_stats.index, rotation=0, fontsize=14)
ax.tick_params(axis='y', labelsize=14)
ax.legend(title='Toxicity', fontsize=14, title_fontsize=16)
ax.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.savefig("like_count_toxicity_log.png", dpi=600)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

metrics = ['like_count', 'retweet_count', 'reply_count']
ylabels = ['Like Count', 'Retweet Count', 'Reply Count']
colors = ['#4daf4a', '#e41a1c']


country_codes = {
    'Armenia': 'AM', 'Bangladesh': 'BD', 'Catalonia': 'CT', 'China': 'CN',
    'Cuba': 'CU', 'Honduras': 'HN', 'Indonesia': 'ID', 'Iran': 'IR',
    'Mexico': 'MX', 'Russia': 'RU', 'Saudi': 'SA', 'Serbia': 'RS',
    'Spain': 'ES', 'Tanzania': 'TZ', 'Turkey': 'TR', 'UAE': 'AE',
    'Uganda': 'UG', 'Venezuela': 'VE'
}


fig, axes = plt.subplots(3, 1, figsize=(16, 8), dpi=1200)

for i, metric in enumerate(metrics):
    stats = df.groupby(['country_map', 'Overall_toxic'])[metric].mean().unstack()
    stats.columns = ['Non-Toxic', 'Toxic']
    

    stats['country_code'] = stats.index.map(lambda x: country_codes.get(x.strip(), ''))
    stats = stats[stats['country_code'] != '']
    stats.set_index('country_code', inplace=True)
    stats.sort_index(inplace=True)
    
    # Plot
    stats.plot(kind='bar', width=0.7, color=colors, logy=True, ax=axes[i])
    
    axes[i].set_ylabel(ylabels[i], fontsize=14)
    axes[i].set_xticks(range(len(stats)))
    axes[i].set_xticklabels(stats.index, rotation=0, fontsize=12)
    axes[i].tick_params(axis='y', labelsize=12)
    axes[i].legend(title='Toxicity', fontsize=12, title_fontsize=14)
    axes[i].grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.savefig("combined_toxicity_log.png", dpi=1200)
plt.show()
```

```python
from scipy.stats import ttest_ind
user_stats = df.groupby('userid').agg(
    total_tweets=('Tweet_ID', 'count'),
    toxic_tweets=('Overall_toxic', lambda x: x.sum()),  
    avg_like=('like_count', 'mean'),
    avg_retweet=('retweet_count', 'mean'),
    avg_reply=('reply_count', 'mean'),
    avg_total_engagement=('Total_Engagement', 'mean')
).reset_index()
user_stats['is_toxic_user'] = user_stats['toxic_tweets'] > 0
grouped = user_stats.groupby('is_toxic_user')[['avg_like', 'avg_retweet', 'avg_reply', 'avg_total_engagement']].mean()
print(grouped)
```

```python
user_stats = user_stats.dropna()
toxic_users = user_stats[user_stats['is_toxic_user']]['avg_total_engagement']
non_toxic_users = user_stats[~user_stats['is_toxic_user']]['avg_total_engagement']

t_stat, p_val = ttest_ind(toxic_users, non_toxic_users, equal_var=False)
print(f"T-test: t = {t_stat:.2f}, p = {p_val:.4f}")
```

```python
df_clean = df.dropna(subset=['Total_Engagement'])
toxic_tweets = df_clean[df_clean['Overall_toxic'] == True]['Total_Engagement']
non_toxic_tweets = df_clean[df_clean['Overall_toxic'] == False]['Total_Engagement']

t_stat, p_val = ttest_ind(toxic_tweets, non_toxic_tweets, equal_var=False)
print(f"T-test (tweet-level): t = {t_stat}, p = {p_val}")
```
